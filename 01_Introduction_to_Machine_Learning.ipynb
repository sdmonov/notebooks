# Какво е Машинно Обучение (Machine Learning, ML)?

И така, какво е „машинно обучение“? ML всъщност са много неща. Областта е доста обширна и се разширява бързо, като непрекъснато се разделя и подразделя на различни под-специалности и видове машинно обучение.

Може би най-добре може да се обобщи от изявлението, направено от Arthur Samuel още през 1959 г .: *“Machine Learning is the field of study that gives computers the ability to learn without being explicitly programmed.”(„Наука, която изучава способността на компютрите да се обучават за изпълнение на задача без да бъдат програмирани за конкретната задача.")*

И по-наскоро, през 1997 г., Tom Mitchell дава следната по-съвременна дефиниция: *“A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E.” („Казваме, че дадена компютърна програма се обучава чрез опит E за изпълнение на клас от задачи T с мярка за ефективност P тогава, когато ефективността на програмата при изпълнение на задача от класа Т, измерена чрез мярката P, се подобрява с увеличаване на опита E.")*

Така че, ако искате програмата ви да предсказва, например, трафик модели на оживено кръстовище (задача Т), можете да я стартирате чрез алгоритъм за машинно обучение с данни за минали модели на трафик (опит Е) и, ако е успешно „научила”, тогава ще се справи по-добре при прогнозиране на бъдещи модели на трафик (мярка за ефективност P). 

Сложният характер на много реални проблеми обаче често означава, че измислянето на специализирани алгоритми, които да ги решават перфектно всеки път, е непрактично, ако не и невъзможно.

Примери за проблеми, които се решават с машинното обучение включват:

*   „Това рак ли е?“
*   „Каква е пазарната стойност на тази къща?“
*   „Кои от тези хора са добри приятели помежду си?“
*   „Ще избухне ли този ракетен двигател при излитане?”
*   „Ще хареса ли този човек този филм?”
*   „Кой е това?”
*   „Какво каза?”
*   ”Как управляваш това нещо?”

Всички тези проблеми са отлични цели за проект за ML и всъщност ML е прилаган към всеки от тях с голям успех. 

*ML решава проблеми, които не могат да бъдат решени само с цифрови средства.*

В зависимост от ML задачата се оформят два вида обучение:

*   Обучение с учител (надзиравано/контролирано машинно обучение, supervised machine learning) - програмата се „обучава“ по предварително определен набор от „примери за обучение“, които след това улесняват способността й да стигне до точно заключение, когато се предоставят нови данни.
*   Обучение без учител (ненадзиравано машинно обучение, unsupervised machine learning) - програмата получава съвкупност от данни и трябва да намери модели и взаимовръзки в тях.

Основно ще се съсредоточим предимно върху обучението с учител, като на края ще споменем и някои неща за обучението без учител.

# Обучение с учител

В повечето приложения, които използват контролирано обучение, крайната цел е да се разработи фино настроена прогнозна функция $h(x)$ (понякога наричана „хипотеза“). „Обучението“ се състои в използването на усъвършенствани математически алгоритми за оптимизиране на тази функция, така че при дадени входни данни $x$ за определен домейн (да речем, квадратура на къща), тя точно да предскаже някаква добра стойност $h(x)$ (да речем, пазарна цена за споменатата къща).

На практика $x$ почти винаги представлява множество точки от данни. Така например, прогноза за цените на жилища може да вземе впредвид не само квадратура ($x_1$), но и брой спални ($x_2$), брой бани ($x_3$), брой етажи ($x_4$), година на построяване ($x_5$), локация ($x_6$) и т.н. Определянето кои входни данни да се използват е важна част от дизайна на ML. За по-лесно обяснение обаче е най-лесно да приемем, че се използва единична стойност.

Така че, да кажем, че нашият проста прогнозна функция има тази форма:

$h(x)=\theta_0+\theta_1x$,

където $\theta_0$ и $\theta_1$ са константи. Нашата цел е да намерим перфектните стойности на $\theta_0$ и $\theta_1$, за да накараме функцията да работи възможно най-добре.

Оптимизирането на функцията $h(x)$ се извършва с помощта на примери за обучение. За всеки пример за обучение имаме входна стойност $x\_train$, за която предварително е известен съответният резултат $y$. За всеки пример намираме разликата между известната, правилна стойност $y$ и нашата прогнозирана стойност $h(x\_train)$. С достатъчно примери за обучение тези разлики ни дават полезен начин за измерване на „грешката“ на $h(x)$. След това можем да променим $h(x)$, като променим стойностите на $\theta_0$ и $\theta_1$, за да направим функцията „по-малко грешна“. Този процес се повтаря отново и отново, до намиране на най-добрите стойности за $\theta_0$ и $\theta_1$. По този начин прогнозната функция се обучава и е готова да прави реални прогнози.

```python
def h(x, t0, t1):
  return t0 + t1 * x
```

# Примери за Машинно Обучение


Придържаме се към прости проблеми за по лесно разбиране, но причината ML да съществува, е защото в реалния свят проблемите са много по-сложни. Тук можем да покажем картина на най-много триизмерен набор от данни, но проблемите с ML често се справят с данни с милиони измерения и много сложни прогнозни функции. ML решава проблеми, които не могат да бъдат решени само с цифрови средства.

Имайки това предвид, нека разгледаме един прост пример. Да приемем, че имаме следните данни за обучение, при които служителите на компания са оценили своето удовлетворение по скала от 1 до 100:


```python
import pandas as pd
import matplotlib.pyplot as plt

salary      =[38,40,52,54,55,56,57,58,62,62,63,63,64,65,66,68,69,72,72,77,77,79,79,80,84,84,85,87,87,92]
satisfaction=[40,46,42,62,61,42,45,46,62,64,61,64,67,59,62,50,55,62,70,70,75,74,67,77,78,68,70,73,82,80]

def plot_data():
  plt.plot(salary, satisfaction, 'bx')
  plt.xlim([30,100])
  plt.ylim([0,100])
  plt.xlabel("Salary in $1000s")
  plt.ylabel("Employee satisfaction rating")

plot_data()
plt.show()
```

Първо, забележете, че данните са малко шумни. Но можем да видим, че има някакъв модел в данните (т.е. удовлетвореността на служителите има тенденция да се увеличава, докато заплатата се покачва), но не всичко се вписва добре в права линия. Това винаги ще съществува при реални данни (и ние абсолютно искаме да обучим нашата програма, използвайки реални данни!). И така, тогава как можем да обучим една програма, за да предскаже перфектно нивото на удовлетвореност на служителя? Отговорът, разбира се, е, че не можем. *Целта на ML никога не е да правите „перфектни“ предположения, защото ML се занимава с домейни, където няма такова нещо. Целта е да се правят предположения, които са достатъчно добри, за да бъдат полезни.*
И така, нека да дадем на нашата програма данните, които сме получили по-горе, и нека тя да ги научи. Първо трябва да инициализираме нашата функция $h(x)$ с някои разумни стойности на $\theta_0$ и $\theta_1$. Сега нашата функция изглежда така, когато е поставена върху нашите тренировъчни данни:

$h(x)=12.0+0.20x$

```python
t0,t1 = 12.0,0.2
y_data=[h(salary[i],t0,t1) for i in range(len(salary))]

plot_data()
plt.plot(salary, y_data, 'r-')
plt.show()
```

Ако попитаме тази прогнозна функция за удовлетворението на служител, който прави $60k, той ще предскаже рейтинг от 24:

```python
r = h(60,12.0,0.2)

plot_data()
plt.plot(salary, y_data, 'r-')
plt.plot([60],[r], 'rx')
plt.show()
```

Очевидно е, че това беше ужасно предположение и че тази програма не знае много много.

Така че сега, нека дадем на тази функция всички заплати от нашия набор от обучения и да вземем разликите между получените прогнозни оценки на удовлетвореност и действителните оценки на удовлетвореността на съответните служители. Ако изпълним малко математическо магьосничество (което ще обясним скоро), можем да изчислим с голяма точност, че ако $\theta_0 = 13.12$ и $\theta_1=0.61$ ще получим по-добра прогнозна функция:

$h(x)=13.12+0.61x$

```python
t0,t1 = 13.12,0.61
y1_data=[h(salary[i],t0,t1) for i in range(len(salary))]

r = h(60,t0,t1)

plot_data()
plt.plot(salary, y1_data, 'r-')
plt.plot([60],[r], 'rx')
plt.plot(salary, y_data, 'm-')
plt.show()
```

И ако повторим този процес, да речем 1500 пъти, нашата функция ще изглежда така:

$h(x)=15.54+0.75x$

```python
t0,t1 = 15.54,0.75
y_best=[h(salary[i],t0,t1) for i in range(len(salary))]

t_test=[(15.5,0.73),(15.3,0.70),(13.12,0.61),(12.0,0.2)]

plot_data()
plt.plot(salary, y_best, 'r-')
for t0_,t1_ in t_test:
  plt.plot(salary, [h(salary[i],t0_,t1_) for i in range(len(salary))], 'm-')
plt.show()
```

На този етап, ако повторим процеса, ще открием, че $\theta_0$ и $\theta_1$ вече няма да се променят съществено и по този начин виждаме, че системата е оптимална. Ако не сме допуснали грешки, това означава, че сме намерили оптималната прогнозна функция. Съответно, ако сега отново попитаме машината за оценка на удовлетвореността на служителя, който прави $60k, тя ще предвиди рейтинг от около 60.

```python
r = h(60,t0,t1)

plot_data()
plt.plot(salary, y_best, 'r-')
plt.plot([60],[r], 'rx')
plt.show()
```

# Регресия в машинното обучение: бележка за сложността 

Горният пример е технически прост проблем на едномерната линейна регресия, който в действителност може да бъде решен чрез извеждане на обикновено нормално уравнение и изцяло пропускане на този „настройващ“ процес. Помислете обаче за прогнозна функция, която изглежда така:

$h(x_1,x_2,x_3,x_4)=\theta_0+\theta_1x_1+\theta_2x_2^2+\theta_3x_3^3x_4+\theta_4x_1^3x_2^2+\theta_5x_2^4x_3x_4^2$

Тази функция вход в четири измерения и има различни полиноми. Извеждането на нормално уравнение за тази функция е значително предизвикателство. *Много съвременни проблеми с машинното обучение отнемат хиляди или дори милиони измерения на данни, за да изградят прогнози, използвайки стотици коефициенти.* Предсказването как ще се изрази геномът на организма или какъв ще е климатът след петдесет години, са примери за такива сложни проблеми.

За щастие итеративният подход, възприет от системите за ML, е много по-издръжлив пред такава сложност. Вместо да използва груба сила, системата за машинно обучение „усеща пътя си” към отговора. При големи проблеми това работи много по-добре. Въпреки че това не означава, че ML може да реши всички произволно сложни проблеми (не може), той остава невероятно гъвкав и мощен инструмент.

# Градиентно спускане - Минимизиране на „грешността“ 

Нека разгледаме по-отблизо как работи този итеративен процес. От горния пример как да се уверим, че $\theta_0$ и $\theta_1$ се подобряват с всяка стъпка, а не по-лошо? Отговорът се крие в „измерването на грешката“, споменато по-рано, заедно с малко смятане.

Измерването на грешка е известно като **функция на цената** (също известна и като **функция на загубата**), **$J(\theta)$**. Входа $\theta$ представлява всички коефициенти, които използваме в нашата прогнозна функция. В нашия случай $\theta$ дефакто е двойката $\theta_0$ и $\theta_1$. $J(\theta_0,\theta_1)$ ни дава математическо измерване на това колко грешенa е нашата прогнозна функция, когато използва дадените стойности на $\theta_0$ и $\theta_1$.

Изборът на функцията на цена е друга важна част от ML програмата. В различни контексти „грешен“ може да означава много различни неща. В нашия пример за удовлетвореност на служителите добре установеният стандарт е линейната функция на най-малките квадрати:

$J(\theta_0,\theta_1)=\frac{1}{2m}\sum_{i=1}^{m} (h(x_{t,i})-y)^2$

При най-малките квадрати наказанието за лошо предположение нараства квадратично с разликата между предположението и верния отговор, така че действа като много „строго“ измерване на грешката. Функцията на цената изчислява средно наказание за всички примери за обучение.

Сега се вижда, че нашата цел е да намерим оптимални $\theta_0$ и $\theta_1$ за нашата функция $h(x)$, така че функцията на цената $J(\theta_0,\theta_1)$ да е възможно най-малка.

Да си представим следната функция на цена за определен проблем за машинното обучение:

Тук можем да видим цената, свързана с различни стойности на $\theta_0$ и $\theta_1$. Също се забелязва, че графиката има куполовидна форма. Дъното на купата представлява най-ниската цена, която може да ни даде прогнозната функция въз основа на дадените данни за обучение. Целта е да се „търкулните надолу по хълма“ и да намерите $\theta_0$ и $\theta_1$, съответстващи на тази точка.

Това е мястото, където математическия анализ влиза в този урок за машинно обучение. За да не усложняваме нещата, няма да изписваме уравненията тук, но по същество това, което правим, е да вземем градиента на $J(\theta_0,\theta_1)$, който е двойката производни на $J(\theta_0,\theta_1)$ ($\frac{1}{\theta_0}$ и $\frac{1}{\theta_1}$). Градиентът ще бъде различен за всяка различна стойност на $\theta_0$ и $\theta_1$ и ни казва какъв е „наклонът на хълма“ и по-специално „кой път е надолу“ за тези конкретни $\theta$. Например, когато включим нашите текущи стойности на $\theta$ в градиента, това може да ни каже, че добавянето на малко към $\theta_0$ и изваждането на малко от $\theta_1$ ще ни отведе по посока на функцията за цена - дъното на долината. Следователно, добавяме малко към $\theta_0$ и изваждаме малко от $\theta_1$ и voilà! Завършихме един кръг от нашия учебен алгоритъм. Нашата актуализирана прогнозна функция, $h(x)=\theta_0+\theta_1x$, ще прави по-добри прогнози от преди. Нашата програма вече е малко по-умна.

Този процес на редуване между изчисляване на текущия градиент и актуализиране на $\theta$ от резултатите е известен като градиентно спускане.

# Класификации в машинното обучение
